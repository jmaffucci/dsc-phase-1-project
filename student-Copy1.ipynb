```python
## Remove commas from dollar columns

tables['tn_movie_budgets']['production_budget'].replace(',','', regex=True, 
                                                        inplace=True)
tables['tn_movie_budgets']['domestic_gross'].replace(',','', regex=True, 
                                                     inplace=True)
tables['tn_movie_budgets']['worldwide_gross'].replace(',','', regex=True, 
                                                     inplace=True)
## Remove dollar signs from same columns

tables['tn_movie_budgets']['worldwide_gross'] \
            = tables['tn_movie_budgets']['worldwide_gross'].str.replace('$','')
tables['tn_movie_budgets']['domestic_gross'] \
            = tables['tn_movie_budgets']['domestic_gross'].str.replace('$','')
tables['tn_movie_budgets']['production_budget'] \
            = tables['tn_movie_budgets']['production_budget'].str.replace('$','')

## Change data type to integer for same columns

tables['tn_movie_budgets']['production_budget'] \
                =tables['tn_movie_budgets']['production_budget'].astype(int)
tables['tn_movie_budgets']['domestic_gross'] \
                    =tables['tn_movie_budgets']['domestic_gross'].astype(int)
tables['tn_movie_budgets']['worldwide_gross'] \
                    =tables['tn_movie_budgets']['worldwide_gross'].astype(int)
```
