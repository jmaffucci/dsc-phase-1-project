```python
tables = {}
dashes = '---'*25

for file in files:
    table_name = file.replace(".csv.gz", '').split('/')[-1].replace('.','_')
    print(dashes)
    
    print(f"Preview of {table_name}")
    tables[table_name] = pd.read_csv(file)
    display(tables[table_name].head(5))
    print()
    
### Source: James Irving
```

```python
### Merge imdb_title_basics with imdb_title_ratings

titles_and_ratings = pd.merge(tables['imdb_title_basics'], 
                                                 tables['imdb_title_ratings'])

### Merge bom_movie_gross with tn_movie_budgets 

financials = pd.merge(tables['bom_movie_gross'], tables['tn_movie_budgets'], 
                                          left_on='title', right_on='movie')
```

```python
financials.drop(['domestic_gross_x', 'foreign_gross', 'domestic_gross_y', 'release_date', 'id', 'studio', 'movie'], 
                                                        axis=1, inplace=True)
del titles_and_ratings['original_title']
```

```python
df = pd.merge(financials, titles_and_ratings, left_on='title', right_on='primary_title')
```

```python
df.drop_duplicates(subset=['title'], inplace=True)
df.drop(['start_year', 'tconst', 'primary_title'], axis=1, inplace=True)
```

```python
df.dropna(axis=0, subset=['genres', 'runtime_minutes'], inplace=True)
```

```python
display(df.info())
df.head(2)
```

```python
## Remove commas from financial metric columns

df['production_budget'].replace(',','', regex=True, inplace=True)
df['worldwide_gross'].replace(',','', regex=True, inplace=True)

## Remove dollar signs from financial metric columns

df['production_budget'] = df['production_budget'].str.replace('$','')
df['worldwide_gross'] = df['worldwide_gross'].str.replace('$','')

## Change data type to integer for financial columns

df['production_budget'] = df['production_budget'].astype(int)
df['worldwide_gross'] = df['worldwide_gross'].astype(int)
```

```python
df['Profit'] = (df['worldwide_gross'] - df['production_budget'])
df['ROI'] = (df['Profit'] / df['production_budget'])
```

```python
df = df[['title', 'year', 'worldwide_gross', 'production_budget','Profit',
         'ROI', 'runtime_minutes', 'genres', 'averagerating','numvotes']]
```
