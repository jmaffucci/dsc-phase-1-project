```python
files = glob.glob(f"{folder}*.csv*")
files
```

```python
files[0].replace(".csv.gz", '').split('/')[-1].replace('.','_')
```

```python
tables = {}
dashes = '---'*25

for file in files:
    table_name = file.replace(".csv.gz", '').split('/')[-1].replace('.','_')
    print(dashes)
    
    print(f"Preview of {table_name}")
    tables[table_name] = pd.read_csv(file)
    display(tables[table_name].head(10))
    print()
```

```python
display(tables['imdb_title_basics'].info())
display(tables['imdb_title_ratings'].info())
display(tables['bom_movie_gross'].info())
display(tables['tn_movie_budgets'].info())
```

```python
display(tables['imdb_title_basics'].describe())
display(tables['imdb_title_ratings'].describe())
display(tables['bom_movie_gross'].describe())
display(tables['tn_movie_budgets'].describe())
```

```python
tables['imdb_title_basics_w_ratings'] = pd.merge(tables['imdb_title_basics'], 
                                                 tables['imdb_title_ratings'])
tables['imdb_title_basics_w_ratings']
```

```python
tables['tn_movie_budgets']['production_budget'].replace(',','', regex=True, 
                                                        inplace=True)
tables['tn_movie_budgets']['production_budget']
```

```python
tables['tn_movie_budgets']['domestic_gross'].replace(',','', regex=True, 
                                                     inplace=True)
tables['tn_movie_budgets']['domestic_gross']
```

```python
tables['tn_movie_budgets']['domestic_gross'] \
            = tables['tn_movie_budgets']['domestic_gross'].str.replace('$','')
tables['tn_movie_budgets']['production_budget'] \
            = tables['tn_movie_budgets']['production_budget'].str.replace('$','') 
tables['tn_movie_budgets']
```

```python
display(tables['imdb_title_basics_w_ratings'].isna().sum())
display(tables['tn_movie_budgets'].isna().sum())
display(tables['bom_movie_gross'].isna().sum())
```
